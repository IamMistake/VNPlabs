### Requirements

Import libraries and modules that you are going to use

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Send HTTP request to the target Website

```python
url = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/"
response = requests.get(url)
```

check the response status code

```python
response.status_code
```

### Parse the response content with BeautifulSoap

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

### Extract data from the BeautifulSoap object using any selectors, attribute identifiers, etc.

* Product title (selector '.wd-entities-title')
* Product regular price (selector '.woocommerce-Price-amount')
* Product discount price (if available), same selector as regular price
* URL to the product page
* Add to cart button URL

```python
titles = soup.select_one('.products')
products = titles.select(".product")
nesta = []
for product in products:
  nesto = {}
  nesto['title'] = product.select('.wd-entities-title')[0].find_all('a')[0].text
  prices = product.select('.price')[0].find_all('bdi')
  nesto['reguralPrice'] = int(prices[0].text.replace(".", "").replace("\xa0", ""))
  if len(prices) > 1:
    nesto['discountPrice'] = int(prices[1].text.replace(".", "").replace("\xa0", ""))
  nesto['pageUrl'] = product.find_all('a')[0]['href']
  nesto['btnUrl'] = url + product.select('.wd-add-btn')[0].find_all('a')[0]['href']

  nesta.append(nesto)
nesta
```

Repeat the extraction process for each page of products

```python
def products_info(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    titles = soup.select_one('.products')
    products = titles.select(".product")
    nesta = []
    for product in products:
        nesto = {}
        nesto['title'] = product.select('.wd-entities-title')[0].find_all('a')[0].text
        prices = product.select('.price')[0].find_all('bdi')
        nesto['reguralPrice'] = int(prices[0].text.replace(".", "").replace("\xa0", ""))
        if len(prices) > 1:
            nesto['discountPrice'] = int(prices[1].text.replace(".", "").replace("\xa0", ""))
        nesto['pageUrl'] = product.find_all('a')[0]['href']
        nesto['btnUrl'] = url + product.select('.wd-add-btn')[0].find_all('a')[0]['href']

        nesta.append(nesto)

    return nesta
```

```python
productss = []
url1 = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/"
for i in range(1, 15):
  tmp = url1 + f"{1}/"
  productss += products_info(tmp)
productss
```

### Create a pandas DataFrame with the scraped products

```python
df = pd.DataFrame(productss)
df
```

Save the dataframe as `.csv`
